## What is Pandas ?
  
Pandas is a software library written for the python programming language
for data **manipulation and analysis.**

* Pandas is built on top of the NumPy package, meaning a lot of the structure of NumPy is used or replicated in Pandas.
* Data in pandas is often used to feed statistical analysis in SciPy, plotting functions from Matplotlib, and machine learning algorithms in Scikit-learn.
* The primary two components of pandas are the **Series and DataFrame**

## 5.Select multiple columns in a Pandas Dataframe

* 5.1 **dataframe['column_name']:** used to select single column from dataframe.
* 5.2 **dataframe['column_name_1', 'column_name_2']:** used to select multiple columns from dataframe.
* 5.3 **dataframe[dataframe.columns[1:4]]:** used to select column 1 - column 3 from dataframe.
* 5.4 **dataframe.loc[1:3, ['column_name_1', 'column_name_2']]:** used to select multiple columns and rows from dataframe  based on rows range and columns name.
* 5.5 **dataframe.loc[1:3, 'column_name_1' :  'column_name_2']:** used to select multiple columns and rows from dataframe based on rows and column range.

## 8.Indexing and Selecting Data with Pandas

* 8.1 **Dataframe[]:** This function also known as indexing operator.
* 8.2 **Dataframe.loc[]:** This function is used for labels. If label is being passed as integer then last element will be as mentioned and not (stop -1).
* 8.3 **Dataframe.loc[rows_index_label, columns]:** This function is used for labels. 
    * It can have range of rows label OR a particular rows label.
    * It can have range of column name OR a particular column name.
* 8.4 **Dataframe.iloc[]:** This function is used for positions or integer based and last element will be as stop-1.
* 8.5 **Dataframe.iloc[rows_index, columns]:** This function is used for positions or integer based.
    * It can have range of rows index OR a particular rows index.
    * It can have range of column name/index OR a particular column name/index.

## 9.Boolean Indexing in Pandas Dataframe
* 9.1 **Dataframe.loc[]:** Boolean index can be accessed by .loc[].
* 9.2 **Dataframe.iloc[]:** Boolean index can not be accessed by .iloc[]. Because it will have **TypeError: Cannot index by location index with a non-integer key.**

```python
dict = {'Name': ['Name_1', 'Name_2', 'Name_3', 'Name_4', 'Name_5'],
             'Age':[27, 24, 22, 32,26],
             'Address':['Delhi', 'Bangalore', 'Chennai', 'Pune', 'Noida'],
             'Qualification': ['BCA', 'MCA', 'BSC', 'MSC', 'B.Tech']}
  
df = pd.DataFrame(dict, index = [True, False, True, False,True])
df
```

### Accessing a Dataframe with a boolean index using .loc[]

```python
df.loc[True]
```

```python
df.loc[False]
```

### Accessing a Dataframe with a boolean index using .iloc[]

* **TypeError:** Cannot index by location index with a non-integer key

## 10.Applying a boolean mask to a dataframe

* 10.1 It will only returns those rows, which have **True value in mask.**

```python
dict = {'Name': ['Name_1', 'Name_2', 'Name_3', 'Name_4', 'Name_5'],
             'Age':[27, 24, 22, 32,26],
             'Address':['Delhi', 'Bangalore', 'Chennai', 'Pune', 'Noida'],
             'Qualification': ['BCA', 'MCA', 'BSC', 'MSC', 'B.Tech']}
  
df = pd.DataFrame(dict, index = [0, 1, 2, 3, 4])
df
```

```python
df[[True, False, True, False, False]]
```

```python
df[[True, True, True, True, True]]
```

```python
df[[False, False, False, False, False]]
```

### Masking data based on column value

```python
dict = {'Name': ['Name_1', 'Name_2', 'Name_3', 'Name_4', 'Name_5'],
             'Age':[27, 24, 22, 32,26],
             'Address':['Delhi', 'Bangalore', 'Chennai', 'Pune', 'Noida'],
             'Qualification': ['BCA', 'MCA', 'BSC', 'MSC', 'B.Tech']}
  
df = pd.DataFrame(dict)
df
```
